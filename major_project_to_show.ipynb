## importing book

```python
import requests
url='https://www.gutenberg.org/files/244/244-0.txt'
response=requests.get(url)

#givngencoding

response.encoding='utf-8'

book_text=response.text
```

```python
book_text[:1000]
```

## stripping content of book between start and end
### means removing header footer metadata

```python

#def remove_header_footer_metadata():
start_marker="*** START OF THE PROJECT GUTENBERG EBOOK"
end_marker="*** END OF THE PROJECT GUTENBERG EBOOK"

start_index=book_text.find(start_marker)+len(start_marker)

end_index=book_text.find(end_marker)

cleaned_text=book_text[start_index:end_index].strip()

print('remove metadata header and footer')
```

```python
cleaned_text
```

## removing extra spaces , non ascii characters  and spcial characters

```python
import re

# Remove extra spaces, special symbols, and non-ASCII characters
# Replace multiple spaces with a single space
cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
# Remove non-ASCII characters
cleaned_text = re.sub(r'[^\x00-\x7F]+', ' ', cleaned_text)
# Remove special symbols except basic punctuation
cleaned_text = re.sub(r'[^\w\s.,!?]', '', cleaned_text)

print("Unwanted characters removed!")
```

```python
cleaned_text=cleaned_text[829:]
```

```python
# Function for cleaning text
def more_clean_text(text):

    # Lowercasing the text
    text=text.lower()

    # Removing comma(,), period(.) and newline character(\n)
    text=re.sub('[,.\n]','',text)

    # Replacing hypen with blank space
    text=re.sub('-',' ',text)

    return text
```

```python
# Cleaning Text
cleaned_text=more_clean_text(cleaned_text)
cleaned_text
```

```python
import spacy

nlp=spacy.load('en_core_web_sm')


doc=nlp(cleaned_text)
```

```python

character_names = set()

for ent in doc.ents:
    if ent.label_ == "PERSON":
        character_names.add(ent.text)

print("Extracted character names:", character_names)
```

```python
#preparing dataset for chabot
```

```python
# Create a dataset of character-specific text
character_dataset = {}

for ent in doc.ents:
    if ent.label_ == "PERSON":  # Filter for person entities (characters)
        character_name = ent.text
        if character_name not in character_dataset:
            character_dataset[character_name] = []
        # Add sentences where the character is mentioned
        for sent in doc.sents:
            if character_name in sent.text:
                character_dataset[character_name].append(sent.text)

# Print the dataset for the first character
first_character = list(character_dataset.keys())[0]
print(f"Dataset for {first_character}:")
for sentence in character_dataset[first_character][:5]:
    print(sentence)
```
